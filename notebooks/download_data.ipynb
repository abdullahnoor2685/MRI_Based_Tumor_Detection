```python
# ============================================================
# DATA ACQUISITION & VERSIONING
# MRI Brain Tumor Detection Dataset
# ============================================================

# DATASET SOURCE
# Name    : Brain Tumor MRI Dataset
# URL     : https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset
# License : CC0 Public Domain
# Size    : ~150MB
# Classes : glioma, meningioma, notumor, pituitary

# MANUAL DOWNLOAD INSTRUCTIONS (Kaggle credentials required)
# 1. Go to https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset
# 2. Click Download
# 3. Unzip the downloaded file
# 4. Place the 4 class folders inside data/raw/:
#       data/raw/glioma/
#       data/raw/meningioma/
#       data/raw/notumor/
#       data/raw/pituitary/
# 5. Run split_data.py to create train/val/test splits

print("Primary Dataset: Brain Tumor MRI Dataset")
print("Source: https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset")
print("License: CC0 Public Domain")
print("Place unzipped contents in data/raw/")
```

```python
import os
from pathlib import Path

raw_dir = Path("data/raw")
classes = ["glioma", "meningioma", "notumor", "pituitary"]
valid_extensions = {".jpg", ".jpeg", ".png", ".bmp"}

print("=" * 50)
print("DATASET VERIFICATION — FILE COUNT PER CLASS")
print("=" * 50)

total = 0
for cls in classes:
    cls_dir = raw_dir / cls
    if not cls_dir.exists():
        print(f"{cls:<15}: MISSING — download dataset first")
        continue
    count = sum(1 for f in cls_dir.iterdir() if f.suffix.lower() in valid_extensions)
    total += count
    print(f"{cls:<15}: {count} images")

print("-" * 50)
print(f"{'TOTAL':<15}: {total} images")
print("=" * 50)

expected = {
    "glioma"     : 1621,
    "meningioma" : 1645,
    "notumor"    : 2000,
    "pituitary"  : 1757,
}
expected_total = sum(expected.values())
print(f"\nExpected total : {expected_total}")
print(f"Found total    : {total}")

if total >= expected_total - 20:
    print("✓ CHECKSUM PASSED — download complete and verified")
else:
    print(f"⚠ CHECKSUM WARNING — check for missing or corrupted files")
```

```python
split_dir = Path("data")
splits = ["train", "val", "test"]

print("=" * 50)
print("SPLIT VERIFICATION")
print("=" * 50)
print(f"{'Class':<15}", end="")
for split in splits:
    print(f"{split:>10}", end="")
print(f"{'Total':>10}")
print("-" * 50)

grand_total = 0
for cls in classes:
    print(f"{cls:<15}", end="")
    row_total = 0
    for split in splits:
        cls_dir = split_dir / split / cls
        if cls_dir.exists():
            count = sum(1 for f in cls_dir.iterdir() 
                       if f.suffix.lower() in valid_extensions)
        else:
            count = 0
        row_total += count
        print(f"{count:>10}", end="")
    grand_total += row_total
    print(f"{row_total:>10}")

print("-" * 50)
print(f"{'TOTAL':<15}", end="")
for split in splits:
    split_total = sum(
        sum(1 for f in (split_dir / split / cls).iterdir()
            if f.suffix.lower() in valid_extensions)
        for cls in classes
        if (split_dir / split / cls).exists()
    )
    print(f"{split_total:>10}", end="")
print(f"{grand_total:>10}")
print("=" * 50)
print("Split ratio: 70% train / 15% val / 15% test | Seed: 42")
```
